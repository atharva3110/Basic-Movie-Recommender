```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("movie_recommender v1/tmdb_5000_credits.csv")
```

```python
movies_df = pd.read_csv("movie_recommender v1/tmdb_5000_movies.csv")
```

```python
credits.head()
```

```python
movies_df.head()
```

```python
print("Credits:",credits.shape)
print("Movies:",movies_df.shape)
```

```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id":"id"})
movies_df_merge = movies_df.merge(credits_column_renamed, on="id")
movies_df_merge.head()
```

```python
movies_cleaned_df = movies_df_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_cleaned_df.head()
```

```python
movies_cleaned_df.info()
```

# Using Weighted Average for each movies's Average Rating

### W = (Rv + Cm) / (v + m)

```python
#calculate all the components based on the above fromula
 
v = movies_cleaned_df['vote_count']
R = movies_cleaned_df['vote_average']
C = movies_cleaned_df['vote_average'].mean()
m = movies_cleaned_df['vote_count'].quantile(0.70)   # movies more that 70 percentile
```

```python
movies_cleaned_df['weighted_avg'] = ((R*v)+(C*m))/(v+m)
```

```python
movies_cleaned_df.head()
```

```python
movie_sorted_ranking = movies_cleaned_df.sort_values('weighted_avg',ascending=False)
movie_sorted_ranking[['original_title','vote_count','vote_average','weighted_avg','popularity']].head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_avg = movie_sorted_ranking.sort_values('weighted_avg',ascending = False)
plt.figure(figsize=(12,6))
axis1 = sns.barplot(x = weight_avg['weighted_avg'].head(10), y = weight_avg['original_title'].head(10))
plt.xlim(6, 10)
plt.title('Best Movies by Average Votes', weight = 'bold')
plt.xlabel('Weighted Avg Score', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.savefig('movie_recommender v1/best_movies.png')
```

```python
popularity = movie_sorted_ranking.sort_values('popularity', ascending = False)
plt.figure(figsize=(12,6))
ax1 = sns.barplot(x = popularity['popularity'].head(10), y = popularity['original_title'].head(10) , data = popularity)
plt.title('Most Popular Movies', weight = 'bold')
plt.xlabel('Popularity Score', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.savefig('movie_recommender v1/best_popular_movies.png')
```

# Recommendation based on scaled weighted avg and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()  #convert/transform the values between 0-1
movie_scaled_df = scaling.fit_transform(movies_cleaned_df[['weighted_avg','popularity']])
movie_normalized_df = pd.DataFrame(movie_scaled_df,columns = ['weighted_avg','popularity']) #convert to df
movie_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weight_avg','normalized_popularity']] = movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score'] = movies_cleaned_df['normalized_weight_avg'] * 0.5 + movies_cleaned_df['normalized_popularity'] * 0.5
movies_scored_df = movies_cleaned_df.sort_values(['score'], ascending = False)
movies_scored_df[['original_title','normalized_weight_avg','normalized_popularity','score']].head(10)
```

```python
scored_df = movies_cleaned_df.sort_values('score', ascending = False)
plt.figure(figsize=(12,6))
ax = sns.barplot(x = scored_df['score'].head(10), y = scored_df['original_title'].head(10) , data = scored_df)
plt.title('Best Popular & Most Rated Movies Blend', weight = 'bold')
plt.xlabel('Score', weight = 'bold')
plt.ylabel('Movie Title', weight = 'bold')
plt.savefig('movie_recommender v1/scored_movies.png')
```
